```python
import pandas as pd
import matplotlib.pyplot as plt

# Upload the CSV file
from google.colab import files
uploaded = files.upload()

# Load the CSV file (adjust the filename if necessary)
df = pd.read_csv('Minoh_room15.csv', index_col=0)

# Clean column names by removing any leading or trailing whitespace
df.columns = df.columns.str.strip()

# Display the first few rows to ensure it loaded correctly
print(df.head())

# Create a time series plot for all columns
fig, axes = plt.subplots(nrows=len(df.columns), ncols=1, figsize=(10, len(df.columns) * 3))
fig.tight_layout(pad=3.0)

# Plot each column
for i, column in enumerate(df.columns):
    axes[i].plot(df.index, df[column])
    axes[i].set_title(f'{column}')
    axes[i].set_xlabel('Time')
    axes[i].set_ylabel(column)

# Show the plots
plt.show()
```
